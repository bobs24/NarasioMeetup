## **2.5. Break and Continue**


*   **Break** digunakan untuk menghentikan secara keseluruhan perulangan
*   **Continue** digunakan untuk melewati perulangan yang sedang terjadi dan melanjutkan ke perulangan berikutnya



### **2.5.1 Contoh Penggunaan Break**

```python
angka = 4
for i in range(10): # Looping dari angka 0 s/d lebih kecil dari 10
    print(i)
    if i is angka:
        print("angka ditemukan",i)
        break # Looping berhenti jika ditemukan angka 4
```

```python
# Contoh tanpa penggunaan Break di source code yang sama
angka = 4
for i in range(10): # Looping dari angka 0 s/d lebih kecil dari 10
    print(i)
    if i is angka:
        print("angka ditemukan",i)
```

### **2.5.2 Contoh Penggunaan Continue**

```python
# inisialisasi list hewan mamalia
mammals = ["dogs", "cats", "horses"]
for x in mammals: # looping tiap item di list mammals
    if x == "cats":
        print("Lanjut looping berikutnya")
        continue # melanjutkan looping berikutnya
    print(x)
```

```python
# Contoh tanpa penggunaan continue di source code yang sama
mammals = ["dogs", "cats", "horses"]
for x in mammals: # looping tiap item di list mammals
    if x == "cats":
      print("Lanjut looping berikutnya")
    print(x)
```

## **2.6. Function**

Anda dapat mengirimkan data yang dikenal sebagai parameter, ke dalam suatu fungsi.

### **2.6.1 Contoh Penggunaan def**

```python
# Membuat fungsi luas segitiga
def luas_segitiga(alas, tinggi): # alas dan tinggi merupakan parameter yang masuk
    luas = (alas * tinggi) / 2
    return luas
```

```python
# Pemanggilan fungsi
# 4 dan 6 merupakan parameter yang diinputkan kedalam fungsi luas segitiga 
luas = luas_segitiga(4, 6)
print("Luas segitiga = ", luas)
```

```python
def my_children(first_name): # function dengan 1 parameter
    print(first_name + " " + "Wijaya")
```

```python
my_children("Theo")
my_children("Tilda")
my_children("Tina")
```

> **Argumen Arbritary untuk menyelamatkan ketika kita tidak tahu caranya banyak argumen yang akan diteruskan ke fungsi Anda,  tambahkan * sebelum nama parameter dalam definisi fungsi.**

```python
def my_children(*first_name): # function dengan multiple parameter
    for nama in first_name:
        print(nama + " " + "Wijaya")
```

```python
my_children("Theo", "Tilda", "Tina")
```

```python
my_children("Andi", "Rudi")
```

```python
# Contoh fungsi barisan fibonacci
def fibonacci(N): 
    L=[]
    a,b=0,1 
    while len(L) < N:
        a,b=b,a+b
        L.append(a) 
    return L
```

```python
fibonacci(10)
```

```python
class instancecounter(object):
    count = 0
    def __init__(self, val):
        self.val = val
        instancecounter.count +=1
    def set_val(self, new_val):
        self.val = new_val
    def get_val(self):
        return self.val
    def get_count(self):
        return instancecounter.count

a = instancecounter(9)
b = instancecounter(18)
c = instancecounter(27)

for obj in (a,b,c):
    print('val of obj: %s' %(obj.get_val()))
    print('count: %s' %(obj.get_count()))
```

```python
# Inheritance
class Date(object):
    def get_date(self):
        return '21-12-2021'
    
class Time(Date):
    def get_time(self):
        return '09:00:00'
    
dt = Date()
print("Get Date from the date class: ", dt.get_date())

tm = Time()
print("Get time: ", tm.get_time())
print("get date from time class by inheriting calling date method: ", tm.get_date())
```

```python
# inheritance
class Animal(object):
    
    def __init__(self, name):
        self.name = name
    def eat(self, food):
        print('%s is eating %s. ' %(self.name, food))
        
class Dog(Animal):
    
    def fetch(self, thing):
        print('%s goes after the %s!' %(self.name, thing))

class Cat(Animal):
    
    def swatstring(self):
        print('%s shreds the string!' %(self.name))
        
d = Dog("Rover")
c = Cat("MeOw")

d.fetch("ball")
c.swatstring()
d.eat("Dog Food")
c.eat("Cat Food")
```

```python
# Sedangkan d.swatstring akan menghasilkan error
# Hal tersebut dikarenakan dalam Dog tidak terdapat atribut swatstring.
d.swatstring()
```

```python
# Polimorfism
class Animal(object):
    
    def __init__(self, name):
        self.name = name
    def eat(self, food):
        print('{0} eats {1}'.format(self.name, food))
        
class Dog(Animal):
    
    def fetch(self, thing):
        print('{0} goes after the {1}!'.format(self.name, thing))
        
    def show_affection(self):
        print('{0} wags tail'.format(self.name))
        
class Cat(Animal):
    
    def swatstring(self):
        print('{0} shreds the string!'.format(self.name))
    
    def show_affection(self):
        print('{0} purrs'.format(self.name))
        
for a in (Dog('Rover'), Cat('Fluffy'), Cat('Precious'), Dog('Scout')): a.show_affection()
```

```python
# Overriding
class thought(object):
    def __init__(self):
        pass
    def message(self):
        print("Thought, always come and go")

class advice(thought):
    def __init__(self):
        super(advice, self).__init__()
    def message(self):
        print("Warning: Risk is always involved when you're dealing with market!")
# advice merupakan subclass yang mengandung thought
```

```python
advice().message()
```

```python
# Inheritance class constructor
import random

class animal(object):
    def __init__(self, name):
        self.name = name
class dog(animal):
    def __init__(self, name):
        super(dog, self).__init__(name)
        self.breed = random.choice(["Doberman", "German Shepard", "Beagle"])
        
    def fetch(self, thing):
        print('%s goes after the %s' %(self.name, thing))

d = dog("dogname")

print(d.name)
print(d.breed)
```

```python
class Point(object):
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    
    # mencari jarak dari titik asal
    def distance(self):
        return (self.x**2 + self.y**2) ** 0.5
```

```python
p1 = Point(6, 8)
jarak = p1.distance()

print(jarak)
```

```python
class Karyawan:
    '''Dasar kelas untuk semua karyawan'''
    jumlah_karyawan = 0

    def __init__(self, nama, gaji):
        self.nama = nama
        self.gaji = gaji
        Karyawan.jumlah_karyawan += 1

    def tampilkan_jumlah(self):
        print("Total karyawan:", Karyawan.jumlah_karyawan)

    def tampilkan_profil(self):
        print("Nama :", self.nama)
        print("Gaji :", self.gaji)
```

```python
# Membuat objek pertama dari kelas Karyawan
karyawan1 = Karyawan("Sarah", 1000000)
# Membuat objek kedua dari kelas Karyawan
karyawan2 = Karyawan("Budi", 2000000)

karyawan1.tampilkan_profil()
karyawan2.tampilkan_profil()
print("Total karyawan :", Karyawan.jumlah_karyawan)
```

```python
class PersegiPanjang:
    def __init__(self,panjang,lebar):
        self.panjang = panjang
        self.lebar = lebar

    def keliling(self):
        return 2 * (self.panjang + self.lebar)

    def luas(self):
        return self.panjang * self.lebar
```

```python
pp = PersegiPanjang(3,4)
keliling_pp = pp.keliling()
luas_pp = pp.luas()
print ('keliling = ',keliling_pp)
print('luas = ',luas_pp)
```
